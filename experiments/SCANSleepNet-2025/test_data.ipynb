The Sleep-EDF (Expanded) dataset is divided into two distinct parts based on where and how the data was collected. Choosing between them changes the nature of your machine learning task:

1. ['cassette'] (Sleep Cassette)
This is the most commonly used subset in research papers.

Context: These were recordings from a study on the effects of age on sleep in healthy individuals (ages 25–101).

Environment: Recorded at the subjects' homes using portable "cassette" recorders.

Data Volume: It contains 153 recordings from 78 subjects.

Use Case: Ideal if you want to train a model to recognize "normal" sleep patterns across a wide age range.

2. ['telemetry'] (Sleep Telemetry)
Context: These recordings come from a study on the effects of a specific sleeping pill (temazepam).

Environment: Recorded in a hospital/lab setting using telemetry (wireless transmission).

Data Volume: Much smaller, containing about 44 recordings from 22 subjects.

Use Case: Used if you are specifically interested in how medication or a clinical environment affects sleep architecture.

```python
from torcheeg.datasets import SleepEDFxDataset
from torcheeg import transforms
from torch.utils.data import DataLoader

# This class automatically downloads (if needed) and processes the files
dataset = SleepEDFxDataset(
    root_path='../../data/sleepEDF/physionet.org/files/sleep-edfx/1.0.0',
    studies=['casstte'], # or 'telemetry'
    channels=['EEG Fpz-Cz'],
    sfreq=100,
    online_transform=transforms.ToTensor(),
    label_transform=transforms.Compose([
        transforms.Select('label'),
        transforms.Mapping({
            'Sleep stage W': 0, 'Sleep stage 1': 1, 'Sleep stage 2': 2,
            'Sleep stage 3': 3, 'Sleep stage 4': 3, 'Sleep stage R': 4
        })
    ]),
    io_path="./.torcheeg/datasets_1773676021141_Nr07b"
)

loader = DataLoader(dataset, batch_size=32, shuffle=True)
```
